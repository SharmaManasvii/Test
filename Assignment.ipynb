# **Fibonacci Series**

```python
%%writefile add.c
 #include <stdio.h>
 int fib(int n) {
    if (n <= 1) {
        return n;
    } else {
        return fib(n - 1) + fib(n - 2);
    }
 }
 int main() {
    int n;
    printf("Enter terms: ");
    scanf("%d", &n);
    for (int i = 0; i < n; i++) {
        printf("%d ", fib(i));
    }
    printf("\n");
    return 0;
}
```

# **Transpose of a Matrix**

```python
 %%writefile tran.c
 #include <stdio.h>
 int main() {

    int a[10][10];
    int n, m;
    printf("Enter rows and columns: ");

    scanf("%d%d", &n, &m);

    printf("Enter elements of the matrix:\n");

    for (int i = 0; i < n; i++) {

        for (int j = 0; j < m; j++) {
            scanf("%d", &a[i][j]);
        }
    }

    printf("Transpose of the matrix:\n");

    for (int i = 0; i < m; i++) {

        for (int j = 0; j < n; j++) {
            printf("%d ", a[j][i]);
        }
        printf("\n");
    }
    return 0;
 }
```

# **Copy file content**

```python
%%writefile copycon.c
#include <stdio.h>

int main() {
    FILE *src, *dest;
    char ch;

    src = fopen("source.txt", "r");
    dest = fopen("copy.txt", "w");

    if (src == NULL || dest == NULL) {
        printf("Error opening file.\n");
        return 1; // Indicate an error
    }

    while ((ch = fgetc(src)) != EOF) {
        fputc(ch, dest);
    }

    printf("File copied successfully.\n");

    fclose(src);
    fclose(dest);

    return 0;
}
```

# **Stack**

```python
%%writefile stack.c

#include <stdio.h>
#include <string.h>
#include <stdlib.h>

#define MAX_SIZE 7

typedef struct {
    char* data[MAX_SIZE];
    int top;
} Stack;

void initialize(Stack* s) {
    s->top = -1;
}

int is_empty(Stack* s) {
    return s->top == -1;
}

int is_full(Stack* s) {
    return s->top == MAX_SIZE - 1;
}

void push(Stack* s, char* day) {
    if (is_full(s)) return;
    s->data[++s->top] = day;
}

char* pop(Stack* s) {
    if (is_empty(s)) return NULL;
    return s->data[s->top--];
}

void print_stack(Stack* s) {
    for (int i = 0; i <= s->top; i++) {
        printf("| %-2s ", s->data[i]);
    }
    for (int i = s->top + 1; i < MAX_SIZE; i++) {
        printf("|    ");   // <-- FIXED LINE
    }
    printf("|\n");
}

void shift_days(Stack* input_stack, Stack* output_stack) {
    while (!is_empty(input_stack)) {
        push(output_stack, pop(input_stack));
    }

    char* shifted_day = pop(output_stack);

    while (!is_empty(output_stack)) {
        push(input_stack, pop(output_stack));
    }

    push(input_stack, shifted_day);
}

int main() {
    char days[MAX_SIZE][3] = {"sa", "su", "mo", "tu", "we", "th", "fr"};

    Stack input_stack, output_stack;
    initialize(&input_stack);
    initialize(&output_stack);

    for (int i = 0; i < MAX_SIZE; i++) {
        push(&input_stack, days[i]);
    }

    printf("--Day Shifting Simulation using two stacks--\n");
    printf("|R# | Sa | Su | Mo | Tu | We | Th | Fr |\n");
    printf("----------------------------------------\n");

    printf("|%-2d ", 1);
    print_stack(&input_stack);

    for (int i = 2; i <= MAX_SIZE + 1; i++) {
        shift_days(&input_stack, &output_stack);
        printf("|%-2d ", i);
        print_stack(&input_stack);
    }

    printf("----------------------------------------\n");

    return 0;
}

```

# **Structure code to count 18+ students**

```python
%%writefile count.c
#include <stdio.h>

struct student {
    char name[50];
    int roll;
    int age;
};

int main() {
    struct student s[10];
    int count = 0;

    // Loop to get input for 10 students
    for(int i = 0; i < 10; i++) {
        printf("Enter name, roll, age of student %d: ", i + 1);
        scanf("%s %d %d", s[i].name, &s[i].roll, &s[i].age);

        if (s[i].age > 18)
            count++;
    }

    printf("\nNumber of students above 18: %d\n", count);

    return 0;
}
```

# **Self refrecial structure code**

```python
%%writefile selfref.c
#include <stdio.h>

struct node {
    int data;
    struct node *next;
};

int main() {
    struct node n1, n2, n3;

    n1.data = 10;
    n2.data = 20;
    n3.data = 30;

    n1.next = &n2;
    n2.next = &n3;
    n3.next = NULL;

    struct node *ptr = &n1;

    printf("Linked list data: ");
    while (ptr != NULL) {
        printf("%d ", ptr->data);
        ptr = ptr->next;
    }

    return 0;
}
```

# **Operations of Matrix**

```python
 %%writefile matrix.c
  #include <stdio.h>

int main() {

    int a[10][10], b[10][10], c[10][10];
    int r, c1;
    int ch;

    printf("Enter rows and columns for matrices: ");
    scanf("%d%d", &r, &c1);

    printf("Enter elements of the first matrix:\n");
    for (int i = 0; i < r; i++) {
        for (int j = 0; j < c1; j++) {
            scanf("%d", &a[i][j]);
        }
    }

    printf("Enter elements of the second matrix:\n");
    for (int i = 0; i < r; i++) {
        for (int j = 0; j < c1; j++) {
            scanf("%d", &b[i][j]);
        }
    }

    printf("Choose an operation:\n");
    printf("1. Addition\n");
    printf("2. Subtraction\n");
    printf("3. Multiplication\n");
    printf("Enter your choice (1, 2, or 3): ");
    scanf("%d", &ch);

    if (ch == 1) {

        // Matrix Addition
        printf("Result of Addition:\n");
        for (int i = 0; i < r; i++) {
            for (int j = 0; j < c1; j++) {
                c[i][j] = a[i][j] + b[i][j];
                printf("%d ", c[i][j]);
            }
            printf("\n");
        }

    } else if (ch == 2) {

        // Matrix Subtraction
        printf("Result of Subtraction:\n");
        for (int i = 0; i < r; i++) {
            for (int j = 0; j < c1; j++) {
                c[i][j] = a[i][j] - b[i][j];
                printf("%d ", c[i][j]);
            }
            printf("\n");
        }

    } else if (ch == 3) {

        // Matrix Multiplication
        printf("Result of Multiplication:\n");
        for (int i = 0; i < r; i++) {
            for (int j = 0; j < c1; j++) {
                c[i][j] = 0;
                for (int k = 0; k < c1; k++) {
                    c[i][j] += a[i][k] * b[k][j];
                }
                printf("%d ", c[i][j]);
            }
            printf("\n");
        }

    } else {
        printf("Invalid choice. Please enter 1, 2, or 3.\n");
    }

    return 0;
}
```

# **Print Using Pointer Array**


```python
%%writefile array_of_pointers.c
#include <stdio.h>

int main() {
    int arr[5];
    int *ptrs[5];

    printf("Enter 5 integers: ");
    for (int i = 0; i < 5; i++)
        scanf("%d", &arr[i]);

    for (int i = 0; i < 5; i++)
        ptrs[i] = &arr[i];  // storing addresses

    printf("Values using pointer array: ");
    for (int i = 0; i < 5; i++)
        printf("%d ", *ptrs[i]);

    return 0;
}
```
